```python
weights = [[]]
m = 2
n = 1
def log_taylor(x):
    return (x - 1) - (x - 1)**2/2 + (x - 1)**3/3 - (x - 1)**4/4 + (x - 1)**5/5
def exp_taylor(x):
    return 1 + x + x**2/2 + x**3/6 + x**4/24 + x**5/120
def create_data():
    cord = []
    y = []
    m = 1.4
    for i in range(100):
        for j in range(100):
            cord.append([i,j])
            if i > m * j:
                y.append(float(1.0))
            else:
                y.append(float(0.0))
    return cord, y
    
for i in range(m):
    weights.append([])
    for j in range(n):
        weights[i].append(float(1))

def sigmoid(x):
    return 1.0 / (1.0 + exp_taylor(-x))

def forward(x):
    output = []
    for i in range(len(x)):
        output.append(0)
        for j in range(len(x[i])):
            output[i] += weights[j][0] * x[i][j]
        output[i] = sigmoid(output[i])
    return output
def BinaryCrossEntropy(y, yhat):
    return -y * log_taylor(yhat) - (1 - y) * log_taylor(1 - yhat)
def derivativeBCE(x):
    return -1 / x + 1 / (1 - x)
def backward(x, y):
    loss = 0
    for i in range(m):
        for j in range(n):
            weights[i][j] -= 0.1 * (x[i] - y[i]) * x[j]
        loss += BinaryCrossEntropy(y[i], x[i])
    return loss/len(x)

def train(x, y):
    for i in range(100):
        out = forward(x)
        print(backward(out, y))

def main():
    x, y = create_data()
    train(x, y)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd
import numpy as np
import sklearn.datasets as datasets
```

```python
# get data
data = datasets.load_iris()
X = data.data
y = data.target
```

```python
# create network
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(4, 3)
        self.fc2 = nn.Linear(3, 3)
        self.softmax = nn.Softmax(dim=1)
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        x = self.softmax(x)
        return x
```

```python
import numpy as np
zeros = np.zeros((3, 3))
#apply sigmoid function to output
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

sig = sigmoid(zeros)
```

```python
# create model
model = Net()
```

```python
# create loss function
criterion = nn.CrossEntropyLoss()

# create optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

```python
# train
for epoch in range(100):
    # convert numpy to tensor
    inputs = torch.from_numpy(X).float()
    targets = torch.from_numpy(y).long()
    
    # forward
    outputs = model(inputs)
    loss = criterion(outputs, targets)
    
    # backward
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # print loss
    print(loss.item())
```
